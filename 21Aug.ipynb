```python
import pandas as pd 
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("spotify.csv")
```

```python
df.head()
```

```python
df.set_index('Date',inplace=True)
```

```python
df.head()
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(data=df)
plt.show()
```

```python
df.index=pd.to_datetime(df.index)
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(data=df,linewidth=2)
plt.title("Daily Global Stream")
plt.show()
```

```python
df1=pd.read_excel("unemployment.xlsx")
```

```python
df1.dtypes
```

```python
plt.figure(figsize=(14,7))
plt.style.use('seaborn-darkgrid')
sns.lineplot(x="Period",y="Unemployed",hue="Gender",data=df1,ci=None)
plt.show()
```

```python
plt.figure(figsize=(14,7))
plt.style.use('seaborn-darkgrid')
sns.lineplot(x="Period",y="Unemployed",hue="Gender",style="Gender",markers=True,dashes=False,data=df1,ci=None)
plt.show()
```
