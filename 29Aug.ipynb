```python
unemployment=pd.read_excel("OneDrive_1_8-27-2020/unemployment.xlsx")
plt.figure(figsize=(7,7))
sns.violinplot(x=unemployment.Age,y=unemployment.Unemployed,palette="Set1",scale="count",cut=0.5)
plt.show()
```

```python
plt.figure(figsize=(16,7))
sns.violinplot(x=unemployment.Age,y=unemployment.Unemployed,palette="Set2",hue=unemployment.Gender)
plt.show()
```

```python
plt.figure(figsize=(16,7))
sns.violinplot(x=unemployment.Age,y=unemployment.Unemployed,palette="Set2",hue=unemployment.Gender,dodge=False)
plt.show()
```

```python
plt.figure(figsize=(16,7))
sns.violinplot(x=unemployment.Age,y=unemployment.Unemployed,palette="Set2",hue=unemployment.Gender,split=True)
plt.show()
```

```python
plt.figure(figsize=(18,9))
sns.violinplot(x=unemployment.Age,y=unemployment.Unemployed,palette="Set3",hue=unemployment.Gender,split=True,inner="quartile")
plt.show()
```

```python
plt.figure(figsize=(18,9))
sns.violinplot(x=unemployment.Age,y=unemployment.Unemployed,palette="Set3",hue=unemployment.Gender,split=True,inner="stick")
plt.show()
```

```python
fig1,axes=plt.subplots(nrows=2,ncols=2,figsize=(20,20))
sns.violinplot(x='species',y="petal_width",ax=axes[0,0],data=iris,size=8,inner=None)
sns.violinplot(x='species',y="petal_length",ax=axes[0,1],data=iris,size=8,inner=None)
sns.violinplot(x='species',y="sepal_width",ax=axes[1,0],data=iris,size=8,inner=None)
sns.violinplot(x='species',y="sepal_length",ax=axes[1,1],data=iris,size=8,inner=None)
plt.show()
```

```python
fig1,axes=plt.subplots(nrows=2,ncols=2,figsize=(20,20))
sns.violinplot(x='species',y="petal_width",ax=axes[0,0],data=iris,size=8,inner=None)
sns.swarmplot(x='species',y="petal_width",ax=axes[0,0],data=iris,size=8,color="white")
sns.violinplot(x='species',y="petal_length",ax=axes[0,1],data=iris,size=8,inner=None)
sns.swarmplot(x='species',y="petal_length",ax=axes[0,1],data=iris,size=8,color="white")
sns.violinplot(x='species',y="sepal_width",ax=axes[1,0],data=iris,size=8,inner=None)
sns.swarmplot(x='species',y="sepal_width",ax=axes[1,0],data=iris,size=8,color="white")
sns.violinplot(x='species',y="sepal_length",ax=axes[1,1],data=iris,size=8,inner=None)
sns.swarmplot(x='species',y="sepal_length",ax=axes[1,1],data=iris,size=8,color="white")
plt.show()
```

```python
plt.figure(figsize=(11,9))
sns.catplot(x='Age',y='Unemployed',col="Gender",kind="violin",palette="Set3",height=8,aspect=1,scale="count",data=unemployment)
plt.show()
```
