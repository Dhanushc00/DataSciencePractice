```python
#Plotting multiple sets of data
x1= [1,3,5,7] 
x2=[2,4,6,8]
y1 = [7,7,7,7]
y2= [17,18,29,40]
plt.figure(figsize=(8,6))
ax = plt.axes()
ax.set_facecolor("white")
plt.bar(x1,y1,label = "First",color = '#42B300') 
# First set of data
plt.bar(x2,y2,label = "Second",color = '#94E413') 
# Second set of data
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.title ('$Bar $ $ Chart$')
plt.legend()
plt.show()
```

```python
# Horizontal Bar Chart
Age = [28,33,43,45,57]
Name = ["Asif", "Steve", 'John', "Ravi", "Basit"]
plt.barh(Name,Age, color ="yellowgreen")
plt.show()
```

```python
# Changing the width of Bars
num1 = np.array([1,3,5,7,9])
num2 = np.array([2,4,6,8,10])
plt.figure(figsize=(8,4))
plt.bar(num1, num1**2, width=0.2 , color = '#FF6F00')
plt.bar(num2, num2**2, width=0.2 , color = '#FFB300')
plt.plot()
```

```python
# Changing the width of Bars
num1 = np.array([1,3,5,7,9])
num2 = np.array([2,4,6,8,10])
plt.figure(figsize=(8,4))
plt.bar(num1, num1**2, width=0.3 , color = '#3C6F00')
plt.bar(num2, num2**2, width=0.3, color = '#8DB300')
plt.plot()
for x,y in zip(num1,num1**2):
      plt.text(x, y-5, '%d' % y, ha='center' , va= 'bottom')
for x,y in zip(num2,num2**2):
      plt.text(x, y+0.05, '%d' % y, ha='center' , va= 'bottom')
plt.plot()
```

```python
plt.style.use('seaborn-darkgrid') 
x1= ['Asif','Basit','Ravi','Minil'] 
y1= [17,18,29,40] 
y2 = [20,21,22,23] 
y2_new=np.add(y1,y2) 
plt.figure(figsize=(8,5)) 
plt.barh(x1,y1,label = "Open Tickets",color = '#FF6F00') 
plt.barh(x1,y2,label = "Closed Tickets", left = y1 , color = '#FFB300') 
plt.xlabel('$X$') 
plt.ylabel('$Y$') 
plt.title ('$Bar $ $ Chart$') 
plt.legend() 
for i, v in enumerate(y1):     
    plt.text(v-5, i, str(v),ha='left', color='blue', fontweight='bold') 
for i, v in enumerate(y2_new):     
    plt.text(v-5, i, str(v), ha='left',color='green', fontweight='bold') 
plt.plot() 
plt.show()
```

```python
plt.style.use('seaborn-darkgrid')
x1= ['Asif','Basit','Ravi','Minil']
y1= [17,18,29,40]
y2 = [20,21,22,23]
y2_new=np.add(y1,y2)
plt.figure(figsize=(8,5))
plt.barh(x1,y1,label = "Open Tickets",color = '#FF6F00')
plt.barh(x1,y2,label = "Closed Tickets", left = y1 , color = '#FFB300')
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.title ('$Bar $ $ Chart$')
plt.legend()
for i, v in enumerate(y1):
    plt.text(v-5, i, str(v),ha='left', color='blue', fontweight='bold')
for i, v in enumerate(y2_new):
    plt.text(v-5, i, str(v), ha='left',color='green', fontweight='bold')
plt.plot()
plt.show()
```
