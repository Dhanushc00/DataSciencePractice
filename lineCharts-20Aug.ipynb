```python
plt.figure(figsize=(10,5))
x=np.linspace(0,10,40)
y1=np.power(x,2)
y2=np.power(x,3)
plt.plot(x,y1,'b-',label='$y1=x^2$')
plt.plot(x,y2,'r-',label='$y2=x^3$')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
x=np.linspace(0,10,2000)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1,'b-',label='sin(x)')
plt.plot(x,y2,'r-',label='$cos(x)$')
plt.xlabel(r'$X$',fontsize=18)
plt.ylabel(r'$Y$',fontsize=18,rotation=-20)
plt.title("$Sin(x) $ $ & $ $ Cos(x)$",fontsize=14)
plt.legend(loc='lower left')
plt.show()
```

```python
plt.figure(figsize=(14,6))
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.subplot(1,2,2)
plt.plot(x,y2)
plt.show()
```

```python
x=np.arange(-50,50)
y1=np.power(x,2)
y2=np.power(x,3)
y3=np.sin(x)
y4=np.cos(x)
y5=np.tan(x)
y6=np.tanh(x)
y7=np.sinh(x)
y8=np.cosh(x)
y9=np.exp(x)

fig,ax1=plt.subplots(nrows=3,ncols=3,figsize=(20,20))
ax1[0,0].plot(x,y1,"tab:blue")
ax1[0,0].set_title("Square Function")
ax1[0,0].set_xlabel(r'$X$',fontsize=18)
ax1[0,0].set_ylabel(r'$Y$',fontsize=18)

ax1[0,1].plot(x,y2,"tab:orange")
ax1[0,1].set_title("Cubic Function")
ax1[0,1].set_xlabel(r'$X$',fontsize=18)
ax1[0,1].set_ylabel(r'$Y$',fontsize=18)

ax1[0,2].plot(x,y3,"tab:green")
ax1[0,2].set_title("Hyperbolic Function")
ax1[0,2].set_xlabel(r'$X$',fontsize=18)
ax1[0,2].set_ylabel(r'$Y$',fontsize=18)

ax1[1,0].plot(x,y4,"tab:purple")
ax1[1,0].set_title("Cubic Function")
ax1[1,0].set_xlabel(r'$X$',fontsize=18)
ax1[1,0].set_ylabel(r'$Y$',fontsize=18)

ax1[1,1].plot(x,y5,"tab:orange")
ax1[1,1].set_title("Cubic Function")
ax1[1,1].set_xlabel(r'$X$',fontsize=18)
ax1[1,1].set_ylabel(r'$Y$',fontsize=18)


ax1[1,2].plot(x,y6,"tab:orange")
ax1[1,2].set_title("Cubic Function")
ax1[1,2].set_xlabel(r'$X$',fontsize=18)
ax1[1,2].set_ylabel(r'$Y$',fontsize=18)

ax1[2,0].plot(x,y7,"tab:orange")
ax1[2,0].set_title("Cubic Function")
ax1[2,0].set_xlabel(r'$X$',fontsize=18)
ax1[2,0].set_ylabel(r'$Y$',fontsize=18)

ax1[2,1].plot(x,y8,"tab:orange")
ax1[2,1].set_title("Cubic Function")
ax1[2,1].set_xlabel(r'$X$',fontsize=18)
ax1[2,1].set_ylabel(r'$Y$',fontsize=18)

ax1[2,2].plot(x,y9,"tab:orange")
ax1[2,2].set_title("Cubic Function")
ax1[2,2].set_xlabel(r'$X$',fontsize=18)
ax1[2,2].set_ylabel(r'$Y$',fontsize=18)
```

```python
import seaborn as sns
import pandas as pd 
import numpy as np 
```

```python
col1=np.linspace(0,10,1000)
col2=np.sin(col1)
df=pd.DataFrame({"C1":col1,"C2":col2})
df.head()
```

```python
plt.style.use('seaborn-darkgrid')
%matplotlib inline
sns.lineplot(x=df.C1,y=df.C2,data=df)
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.set(rc={"axes.facecolor":"#fffde7","axes.grid":True,'xtick.labelsize':14,"ytick.labelsize":14})
sns.lineplot(x=df.C1,y=df.C2,data=df,linewidth=2)
sns.lineplot(x=df.C1,y=df.C2+1,data=df,linewidth=2)
plt.show()
```
